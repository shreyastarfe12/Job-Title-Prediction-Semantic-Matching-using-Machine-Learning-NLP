```python
from sklearn.preprocessing import LabelEncoder

# List of categorical columns to encode
cat_columns = [
    'Experience', 'Qualifications', 'location', 'Country',
    'Work Type', 'Company Size', 'Preference', 'Job Title',
    'Role', 'Job Description', 'Benefits', 'skills',
    'Responsibilities', 'Company'
]

# Dictionary to store label mappings
label_mappings = {}

# Create a LabelEncoder instance
encoder = LabelEncoder()

# Encode each column
for col_name in cat_columns:
    df[col_name] = encoder.fit_transform(df[col_name].astype(str))
    label_mappings[col_name] = dict(zip(range(len(encoder.classes_)), encoder.classes_))
```

```python
from sklearn.model_selection import train_test_split

# Separate the target column from the predictors
features = df.drop(columns=['Job Title'])
target = df['Job Title']

# Split the data into training and testing sets (80% train, 20% test)
X_train_set, X_test_set, y_train_set, y_test_set = train_test_split(
    features, target, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the random forest classifier with all CPU cores
rf_classifier = RandomForestClassifier(n_jobs=-1, random_state=42)

# Train the model using the training data
rf_classifier.fit(X_train_set, y_train_set)
```

```python
from sklearn.metrics import accuracy_score

# Generate predictions on the test set
y_pred = rf_classifier.predict(X_test_set)

# Calculate the accuracy of the model
model_accuracy = accuracy_score(y_test_set, y_pred)

# Display shapes and accuracy
print(f"Actual labels shape: {y_test_set.shape}")
print(f"Predicted labels shape: {y_pred.shape}")
print(f"Model Accuracy: {model_accuracy:.4f}")
```

```python
from fuzzywuzzy import process

def fuzzy_encode(user_input, target_column):
    # Check if the column is in the mapping dictionary
    if target_column not in label_mappings:
        raise KeyError(f"No encoding found for column: {target_column}")

    # Fetch the label-to-code mapping
    mapping_dict = label_mappings[target_column]

    # Create reverse map: string label → integer code
    str_to_code = {str(value): key for key, value in mapping_dict.items()}

    # Find the best fuzzy match for the user's input
    match, score = process.extractOne(user_input, str_to_code.keys())

    # Return the encoded integer corresponding to the best match
    return str_to_code[match]

# Example use
sample_input = "mid level"
encoded_result = fuzzy_encode(sample_input, 'Job Title')
print(f"Encoded value for '{sample_input}': {encoded_result}")
```

```python
import re
import pandas as pd
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk

# Download required NLTK data (run once)
nltk.download('stopwords')
nltk.download('wordnet')

# Example dataframe - replace this with your actual df
df = pd.read_csv('job_descriptions.csv')

# Preprocessing function for job descriptions
def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    # Lowercase
    text = text.lower()

    # Remove non-alphabetical characters (keep spaces)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenize
    tokens = text.split()

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Join tokens back to string
    return ' '.join(tokens)
```

```python
# Apply preprocessing to job descriptions
df['Processed_Job_Description'] = df['Job Description'].apply(preprocess_text)

# Filter out empty descriptions
job_descriptions = df['Processed_Job_Description'].tolist()
job_descriptions = [desc for desc in job_descriptions if desc]

# Vectorize job descriptions
vectorizer = TfidfVectorizer()
job_vectors = vectorizer.fit_transform(job_descriptions)
```

```python
# Function to match user input to job title based on description similarity
def match_job_description(user_input):
    user_input_processed = preprocess_text(user_input)
    user_vector = vectorizer.transform([user_input_processed])
    similarities = cosine_similarity(user_vector, job_vectors)
    most_similar_idx = similarities.argmax()
    return df.iloc[most_similar_idx]['Job Title']
```

```python
# Example usage
user_input = "Looking for a mid-level Python developer with experience in data analysis"
matched_job = match_job_description(user_input)
print(f"Best matched job: {matched_job}")
```
